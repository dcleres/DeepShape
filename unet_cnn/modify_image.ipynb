```python
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
```

```python
def rotate_img(img_path, angle):
    img = cv2.imread(img_path, 0)
    print(img)
    img2 = cv2.imread(img_path, 1)
    print(img2)
    rows, cols = img.shape
    rot_M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    rot_img = cv2.warpAffine(img, rot_M, (cols, rows))
    return rot_img, img
```

```python
def flip_img(img_path, border_id):
    img = cv2.imread(img_path, 0)
    return cv2.flip(img, border_id)
```

```python
flipped = flip_img('../../training/images/satImage_001.png',1)
```

```python
dst, img = rotate_img('../../training/images/satImage_001.png', 30)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
