```python
import time
import matplotlib.pyplot as plt
import numpy as np

voting_input = np.array(voting_input)

def multi_slice_viewer(volume):
    fig, ax = plt.subplots()
    for i in range (0, 99):
        plt.imshow(volume[i])
        time.sleep(0.5)
        plt.show()
        if (i==99):
            i = 0
```
